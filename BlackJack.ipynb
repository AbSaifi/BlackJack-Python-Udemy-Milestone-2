<h><b>BlackJack Udemy Milestone Project 2<br><br>Gameplay</b></h>
<br><br>
    
<p>    
1. Game is between a player and computer<br>
2. Played with deck of cards<br>
3. Goal is to be the first one with sum of card value greater than 17 and less than equal to 21<br>
4. Card values are as below<br>    
 
  &nbsp;&nbsp;&nbsp;&nbsp;All face cards : 10<br>
  &nbsp;&nbsp;&nbsp;&nbsp;Ace            : Either 1 or 11<br>
  &nbsp;&nbsp;&nbsp;&nbsp;Rest           : Same as card value <br>
  
5.Game has 3 main terms<br>

  &nbsp;&nbsp;&nbsp;&nbsp;Hit   : Means drawing a card from deck<br>
  &nbsp;&nbsp;&nbsp;&nbsp;Stand : Stop drawing card from deck<br>
  &nbsp;&nbsp;&nbsp;&nbsp;Bust  : When sum of card value exceeds 21<br>

6.Player can decide a bet amount which should not exceed the chip amount with him<br>
7.After deciding bet, player will deal<br>
8.On deal, both player and computer will get two cards from deck<br>
9.Player 1 will have both cards face up<br>
10.Computer will have one card face up and one face down<br>
11.Player 1 will get the choice to hit until he decides to stand or he go bust<br>
12.Bust means sum of player cards exceeds 21 and he loses te game<br>
13.Hit means drawing an additional card from the deck<br>
14.Stand means player is signalling that he will draw no more cards<br>
15.If player choses to stand, computer will show his card and start drawing from deck until his sum exceeds 17<br>
16.Here wither computer can go bust, or he will have sum between 17 and 21<br>
17.At this point sum of both computer and player will be compared. Whoever is close to 21 wins<br>
18.Winning doubles the bet amount and adds in chip balance of player<br>
19.Losing reduces the bet amount from chip balance of player<br>
20.On deal, both player and computer will get two cards from deck<br>

<h><b>Limitations<br></b></h>

1.Insurance, Split or Double functionality is not present<br>


<b>To start game, execute function start_game()<br>

```python
class Display():
    
    def __init__(self):
        pass
    
    def clear_screen(self):
        from IPython.display import clear_output
        clear_output()
    
    def display_board(self,p_player,p_computer,p_player_list,p_computer_list):
        self.clear_screen()
        player_list_len = len(p_player_list)
        computer_list_len = len(p_computer_list)
        #print('player_list_len : '+str(player_list_len))
        #print('computer_list_len : '+str(computer_list_len))
        v_separator = ','
        v_var = ' '
        print('{0}({1})'.format(p_computer.name,p_computer.sum).center(40))
        
        for i in range(0,computer_list_len):
            if i == computer_list_len-1:
                v_separator = ' '
            #elif i == 0:
             #   print(' '.rjust(10),end="")
            v_var += str(p_computer_list[i])+v_separator
        print('')    
        print(v_var.center(40))    
        print('')
        print('--------------------------------------')
        print('')
        print('{0}({1})'.format(p_player.name,p_player.sum).center(40))
        
        v_separator = ','
        v_var = ''
        for i in range(0,player_list_len):
            if i == player_list_len-1:
                v_separator = ' '
            #elif i == 0:
             #   print(' '.rjust(10),end="")
            v_var += str(p_player_list[i])+v_separator  
        
        print('')    
        print(v_var.center(40))
```

```python
import random
class Deck():
    
    def __init__(self):
        
        self.cardlist = ['2S','3S','4S','5S','6S','7S','8S','9S','10S','JS','QS','KS','AS',
                          '2C','3C','4C','5C','6C','7C','8C','9C','10C','JC','QC','KC','AC',
                          '2D','3D','4D','5D','6D','7D','8D','9D','10D','JD','QD','KD','AD',
                          '2H','3H','4H','5H','6H','7H','8H','9H','10H','JH','QH','KH','AH']
        
        random.shuffle(self.cardlist)
     
    def draw_card(self):
        
        card_number = self.cardlist[0]
        self.cardlist.pop(0)
        
        return card_number
```

```python
class Account():
    
    def __init__(self,acbalance):
        
        self.acbalance = acbalance    

    def incrbalance(self,amount):
        
        self.acbalance += amount
        
        return self.acbalance
    
    def decrbalance(self,amount):
        
        self.acbalance -= amount
        
        return self.acbalance

    def chkbalance(self):
        
        return self.acbalance
```

```python
class Player():
    
    def __init__(self,name,acbalance,sum,betamount):
        
        self.name = name
        #self.acbalance = acbalance        
        self.account = Account(acbalance)
        self.sum = sum
        self.betamount = betamount
        
    def hit(self,deck):
        
        cardnumber = deck.draw_card()
        return cardnumber
```

```python
def start_game():
    
    from IPython.display import clear_output
    def is_valid_number(p_input_number):
        
        try:
            p_input_number = int(p_input_number)
        except:
            print('Invalid number. Please chose valid amount')
            return 'N'
        
        if p_input_number <= 0:
            print('Please chose an amount greater than 0')
            return 'N'
        
        return 'Y'
                    
    def is_valid_YN(p_input_response,p_val1,p_val2):

        if str(p_input_response) != str(p_val1) and str(p_input_response) != str(p_val2):
            print('Invalid input. Please enter {0} or {1}'.format(p_val1,p_val2))
            return 'N'
        else:
            return 'Y'
        
    def parse_and_return_num(p_input):
        
        if p_input[:-1] in ('J','Q','K'):
            return 10
        elif p_input[:-1] == 'A':
            while True:
                choice_for_a = input('Consider 1 or 11 for A?')
                if is_valid_YN(choice_for_a,1,11) == 'N':
                    continue
                else:
                    return int(choice_for_a)
        else:
            return int(p_input[:-1])  
        
    def game_start_func(player,computer):
        
        import time
        
        if player.account.chkbalance() == 0:
            print('Sorry you have no balance left. Please buy some more chips, we hope to see you again!!')
            time.sleep(2)
            return 'N'
        
        
        player.betamount = 0
        computer.betamount = 0
        player.sum = 0
        computer.sum = 0
        deck = Deck()
        #print('Deck object {0}'.format(deck))
    
        #player_list = ['','','','','','','','','','','']
        #computer_list = ['','','','','','','','','','','']
        player_list = []
        computer_list = []
    
    
        while True:
            player.betamount = input('Please enter bet amount')
            if is_valid_number(player.betamount) == 'Y':
                player.betamount = int(player.betamount)
                currentbalance = player.account.chkbalance()
                if player.betamount > currentbalance:
                    print('Bet amount {0} cannot be more than total avaialble balance {1}'.format(player.betamount,currentbalance))
                else:
                    break
            else:
                continue
                
        print('Game Start!!')
        display = Display()
        #print('Display object {0}'.format(display))
        print('Dealing..')
        time.sleep(2)
    
        for i in range(0,2):
            player_hit = player.hit(deck)
            player_list.append(player_hit)
            player.sum += parse_and_return_num(player_hit)

            computer_hit = player.hit(deck)

            if i != 1:
                computer_list.append(computer_hit)
                computer.sum += parse_and_return_num(computer_hit)
            else:
                computer_list.append('XX')

            computer_hit_2 = computer_hit
        
        display.clear_screen()
        #print(computer_list)
        #print(player_list)
        display.display_board(player,computer,player_list,computer_list)
        
        if check_sum(player,computer) == 'N':
            return 'N'
                
        while True:
            player_choice = input('Player, do you want to Hit or Stand. H for Hit, S for Stand')

            if is_valid_YN(player_choice,'H','S') == 'N':
                continue

            elif player_choice == 'H':
                player_hit = player.hit(deck)
                player_list.append(player_hit)
                player.sum += parse_and_return_num(player_hit)
                display.display_board(player,computer,player_list,computer_list)
                
                if check_sum(player,computer) == 'N':
                    return 'N'
                continue
            else:
                computer_list[1] = computer_hit_2
                computer.sum += parse_and_return_num(computer_hit_2)
                display.display_board(player,computer,player_list,computer_list)
                
                if check_sum(player,computer) == 'N':
                    return 'N'
                
                while check_sum(player,computer) == 'Y':
                    
                    computer_hit = player.hit(deck)
                    computer_list.append(computer_hit)
                    computer.sum += parse_and_return_num(computer_hit)
                    display.display_board(player,computer,player_list,computer_list)
                
                return 'N'
                
    def check_sum(p_player,p_computer):
        if p_player.sum > 21:
            print('Player {0} busted. Computer wins!. Player''s updated balance {1}'.format(p_player.name,p_player.account.decrbalance(p_player.betamount)))
            #return 'N'
        elif p_computer.sum > 21:    
            print('Computer busted. Player {0} wins!. Player''s Updated balance {1}'.format(p_player.name,p_player.account.incrbalance(p_player.betamount)))
            #return 'N'
        elif p_computer.sum > 17:
            if p_computer.sum > p_player.sum:
                print('Computer wins!. Player''s Updated balance {0}'.format(p_player.account.decrbalance(p_player.betamount)))
            elif p_computer.sum < p_player.sum:
                print('Player {0} wins!. Player''s Updated balance {1}'.format(p_player.name,p_player.account.incrbalance(p_player.betamount)))
            elif p_computer.sum == p_player.sum:
                print('Its a Draw!!. Player''s money returned, current balance {0}'.format(p_player.account.chkbalance()))
        else:
            return 'Y'
        
        while True:
            keep_playing = input('Do you want to play again. Y or N?')
            if is_valid_YN(keep_playing,'Y','N') == 'N':
                continue
            elif keep_playing == 'N':
                return 'N'
            else:
                if game_start_func(p_player,p_computer) == 'N':
                    return 'N'
    
    print('Welcome to Black Jack!!')
    playername = input('Please enter your Name')
    
    while True:
        
        totalamount = int(input('Please enter total amount of chips with you'))
        if is_valid_number(totalamount) == 'N':
            continue
        else:    
            cont = 'C'
            if totalamount < 100:
                print('Minimum bet amount is 100. Please get more chips')
                
                while True:
                    cont = input('Do you want to increase chips or exit. Y for increase chips, N for exit.')

                    if is_valid_YN(cont,'Y','N') == 'N':
                        continue
                    else:
                        break            
        
        if cont == 'Y':
            continue        
        elif cont == 'C':
            break
        else:
            return
                    
    player1 = Player(playername,totalamount,0,0)
    computer1 = Player('Computer',99999999,0,0)
    
    if game_start_func(player1,computer1) == 'N':
        clear_output()
        print('Thanks for Playing BlackJack!!')
        return
    


```
